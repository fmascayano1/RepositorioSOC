```python
# Seleccionar variables de interÃ©s
variables_interes = [
    # IdentificaciÃ³n de la prueba
    'id_result', 'guid_result', 'ts_result', 'id_platform',
    
    # InformaciÃ³n del dispositivo
    'attr_device_model', 'attr_device_manufacturer', 'attr_device_os_version',
    
    # InformaciÃ³n de la red y SIM
    'attr_sim_operator_common_name', 'attr_network_operator_common_name',
    'attr_network_operator_mcc', 'attr_network_operator_mnc',
    'attr_connection_type_start_string', 'attr_connection_type_end_string',
    
    # UbicaciÃ³n geogrÃ¡fica
    'attr_location_latitude', 'attr_location_longitude', 'attr_location_accuracy_m','attr_place_region','attr_place_subregion',
    
    # Indicadores de calidad de seÃ±al
    'val_download_kbps', 'val_upload_kbps', 'val_latency_min_ms',
    'val_multiserver_latency_ms', 'val_jitter_ms', 'val_traceroute0_latency_ms',
    'val_server_distance_km',
    
    # MÃ©tricas especÃ­ficas de seÃ±al celular
    'val_signal_rsrp_dbm', 'val_signal_rsrq_db', 'val_signal_rssi_dbm', 'val_signal_cqi',

    
    # Uso de NR (National Roaming)
    'is_device_data_roaming_enabled', 'attr_connection_nr_state'
]

df = df[variables_interes]

# Mostrar las primeras filas del DataFrame con las variables seleccionadas
df.head()
```

```python
# Verificar columnas presentes y faltantes
columnas_presentes = [col for col in variables_interes if col in df.columns]
columnas_faltantes = [col for col in variables_interes if col not in df.columns]

print("Columnas presentes:", columnas_presentes)
print("Columnas faltantes:", columnas_faltantes)

# Seleccionar solo las columnas presentes
df = df[columnas_presentes]

# Mostrar las primeras filas del DataFrame con las variables seleccionadas
```
